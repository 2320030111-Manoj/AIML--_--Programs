```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 24, 36, 48, 55]

# Create the bar plot
plt.bar(categories, values, color='blue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 24, 36, 48, 55]

# Create the scatter plot
plt.scatter(x, y, color='red', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [12, 15, 13, 17, 19, 18, 21, 24, 24, 25, 26, 26, 27, 28, 30, 32, 33, 34, 35, 36, 37]

# Create the histogram
plt.hist(data, bins=5, color='green', edgecolor='black')

# Add labels and title
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (for example, exam scores of students from different schools)
data = [
    [65, 70, 75, 80, 85],  # School A
    [60, 62, 68, 73, 78],  # School B
    [55, 60, 65, 70, 75],  # School C
    [70, 72, 77, 82, 90],  # School D
]

# Create the box plot
plt.boxplot(data, labels=['School A', 'School B', 'School C', 'School D'], patch_artist=True)

# Add labels and title
plt.xlabel('Schools')
plt.ylabel('Exam Scores')
plt.title('Comparison of Exam Scores across Schools')

# Show the plot
plt.show()
```
